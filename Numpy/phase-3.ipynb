# Advance Operations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])

print("=== Zomato sales analysis ===")
print(sales_data.shape)
print(sales_data.dtype)
print(sales_data[:, 1:]) # Here the slicing is used like this -> 
                        # [: -> This states how many rows to select (: means that select all the rows), 
                        # 1: -> This states that how many columns to select (1: means that select all the columns till the ending but start from the 1 index)]
```

```python
# Total sales per year

print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=1)
print(yearly_total)

# Minimum sales per restaurants
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# Average sales per restaurants
mean_sales = np.mean(sales_data[:, 1:], axis=1)
print(mean_sales)

# Cummulative sales
cummulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print(cummulative_sales)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cummulative_sales, axis=0))
plt.title("Avearge cummilative sales accross all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## Vector Operations

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

# Vector addition
print("Vector Addition: ", vector1 + vector2)

# Vector Multiplication
print("Vector Multiplication: ", vector1 * vector2)

# Vector dot product
print("Dot Product in Vector: ", np.dot(vector1, vector2))

# Angle of a vector
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
