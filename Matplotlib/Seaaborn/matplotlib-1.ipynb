# Matplotlib Tutorial

```python
# Importing matl=plotlib 
import matplotlib.pyplot as plt
```

## Line Plot

```python
days = [1,2,3,4,5]

water = [2,3,2.5,4,3.5]

plt.plot(days, water, marker='o', color='skyblue', linestyle='--')
plt.xlabel("Days")
plt.ylabel("Water Intakes (Litres)")
plt.title("Daily water Consumptions")
plt.show()
```

```python
# Task 1
days = [1,2,3,4,5,6,7]

steps = [3000,5000,4500,6000,7000,8000,7500]

plt.plot(days, steps, marker='o', color='green')

plt.title("Daily Steps")

plt.xlabel("Day")
plt.ylabel("Steps")

plt.show()
```

```python
# Task 2
exam = [1,2,3,4,5]

marks = [50,58,65,70,85]

plt.plot(exam, marks, marker='o', linestyle='--', color='red')

plt.title("Exam Result")

plt.xlabel("Exam")
plt.ylabel("Marks")

plt.show()
```

## Multiple lines

```python
epochs = [1,2,3,4,5]

train_acc = [60,70,80,85,90]

val_acc = [58,68,75,80,82]

plt.plot(epochs, train_acc, label="Epoch to Train Ratio")

plt.plot(epochs, val_acc, label='Epoch to value Ratio')

plt.legend()

plt.show()
```

```python
# Task 3
days = [1,2,3,4,5]

target = [3,3,3,3,3]

actual = [2.5,3.2,2.8,4,3.5]

plt.plot(days, target, label='Target Water Intake', marker='o', color='blue')

plt.plot(days, actual, label='Actual Water Intake', marker='s', color='green')

plt.title("Water Intake Analysis")
plt.xlabel("Days")
plt.ylabel("Water Intake")

plt.legend()

plt.show()
```

## Bar Chart

```python
# Bar Chart
department = [
    "IT",
    "HR",
    "Sales",
    "Finance"
]

employees = [
    50,
    20,
    35,
    15
]

plt.bar(department, employees)

plt.show()
```

```python
plt.bar(department, employees)

plt.title("Department Wise Employees")

plt.xlabel("Department")

plt.ylabel("Employee Count")

plt.show()
```

```python
plt.bar(
    department,
    employees,
    color='green'
)
```

```python
plt.bar(
    department,
    employees,
    color=[
        'red',
        'green',
        'blue',
        'orange'
    ]
)
```

```python
# Task 10
subjects = [
    "Math",
    "Science",
    "English",
    "History"
]

marks = [
    85,
    90,
    78,
    88
]

plt.bar(subjects, marks, color='blue')
plt.title("Subject Wise Marks")
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

```python
# Task 11
products = [
    "Laptop",
    "Mobile",
    "Tablet",
    "Monitor"
]

sales = [
    120,
    300,
    80,
    150
]

plt.bar(products, sales, color=['green', 'red', 'blue', 'purple'])
plt.title('Product wise sales')
plt.xlabel('Products')
plt.ylabel('Sales')

plt.show()
```

```python
# ML Challenge
class_name = [
    "Spam",
    "Not Spam"
]

count = [
    100,
    900
]

plt.bar(class_name, count, color=['red', 'green'])
plt.title('Spam Analysis')
plt.xlabel('Class')
plt.ylabel('Count')

plt.show()

# Is the dataset balanced?
# No

# Which class dominates?
# Not Spam

# Why this might cause problem for machine learning?
# The is biased towards Not Spam
```

## Subplot

```python
plt.subplot(1,2,1)
plt.hist([10,20,30,40,50])

plt.subplot(1,2,2)
plt.boxplot([10,20,30,40,50])

plt.show()
```

```python
data = [10,20,30,40,50]

plt.subplot(2,2,1)
plt.hist(data)
plt.title("Histogram")

plt.subplot(2,2,2)
plt.boxplot(data)
plt.title("Box Plot")

plt.subplot(2, 2, 3)
plt.boxplot(data)
plt.title('bar')

plt.tight_layout()
plt.show()
```

```python
# Task 13
marks = [
    55,60,65,70,
    75,80,85,90,95
]

plt.subplot(2, 2, 1)
plt.hist(marks, bins=5)
plt.title('Histogram')

plt.subplot(2, 2, 2)
plt.boxplot(marks)
plt.title('BoxPlot')

# Another graph added
plt.subplot(2, 2, 3)
plt.boxplot(marks)
plt.title('Another Boxplot')

plt.tight_layout()
plt.show()
```
