# Seaborn

## Countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
employees = pd.DataFrame({
    'Department': [
        'IT','HR','IT',
        'Finance','HR',
        'IT','Finance',
        'HR','IT'
    ]
})

employees
```

```python
sns.countplot(
    x='Department', # This can be turned into horizontal countplot when names are long by just changing the axis to y
    data=employees,
    color='orange'
)

plt.title('Employees by Department')
plt.show()
```

```python
# Countplot with hue
df = pd.DataFrame({
    'Department': [
        'IT','IT','HR',
        'HR','Finance',
        'Finance','IT'
    ],
    'Gender': [
        'Male','Female',
        'Male','Female',
        'Male','Female',
        'Male'
    ]
})

sns.countplot(
    x='Department',
    data=df,
    hue='Gender'
)

plt.show()
```

```python
# Tasks

df = pd.DataFrame({
    'Department': [
        'IT','HR','IT',
        'Finance','HR',
        'IT','Finance',
        'HR','IT','Sales',
        'Sales','IT'
    ],
    'Gender': [
        'Male','Female',
        'Male','Female',
        'Male','Female',
        'Male','Female',
        'Male','Female',
        'Male','Female'
    ]
})

# Task 1
sns.countplot(x='Department', data=df)
# Task 2
plt.title('Department Distribution')

plt.show()

```

```python
# Task 3 & Task 4 (I have combined task 3 & 4)
sns.countplot(y='Department', data=df, hue='Gender')
plt.show()
```

## Histplot

```python

df = pd.DataFrame({
    'Age': [
        22,23,24,25,25,
        26,27,28,29,30,
        31,32,35,40,45
    ]
})

sns.histplot(df['Age'], kde=True)

plt.show()
```

```python
import numpy as np

ages = np.random.normal(
    loc=30,
    scale=5,
    size=1000
)

sns.histplot(
    ages,
    kde=True
)

plt.show()
```

```python
df = pd.DataFrame({
    'Marks': np.random.normal(70,10,100)
})

sns.histplot(
    df['Marks'],
    kde=True
)

plt.show()
```

## Pairplot

```python
import pandas as pd

df = pd.DataFrame({
    'Age': [22,25,30,35,40],
    'Salary': [25000,30000,45000,60000,75000],
    'Experience': [1,2,5,8,12]
})

import seaborn as sns

sns.pairplot(df)

plt.show()
```

```python
df = pd.DataFrame({
    'Age':[22,25,30,35,40,28,33,45],
    'Salary':[25000,30000,45000,60000,75000,35000,50000,85000],
    'Experience':[1,2,5,8,12,3,6,15],
    'Gender':[
        'Male','Female',
        'Male','Female',
        'Male','Female',
        'Male','Female'
    ]
})

sns.pairplot(
    df,
    hue='Gender'
)

plt.show()
```

## Heatmaps

```python
df = pd.DataFrame({
    'Age':[22,25,30,35,40],
    'Salary':[25000,30000,45000,60000,75000],
    'Experience':[1,2,5,8,12]
})

df.corr()
import seaborn as sns

corr_matrix = df.corr()

sns.heatmap(corr_matrix)

plt.show()
```

```python
sns.heatmap(
    corr_matrix,
    annot=True
)

plt.show()
```

```python
sns.heatmap(
    df.isnull()
)
```

## Regplot

```python
df = pd.DataFrame({
    'TV_Hours':[1,2,3,4,5],
    'Marks':[90,80,70,60,50]
})

sns.regplot(
    x='TV_Hours',
    y='Marks',
    data=df
)

plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    'Study_Hours':[1,2,3,4,5,6,7,8],
    'Marks':[35,42,50,58,65,73,82,90]
})

sns.regplot(
    x='Study_Hours',
    y='Marks',
    data=df,
    scatter_kws={'s':100},
    line_kws={"color":'red'}
)

plt.show()
```

```python
df = pd.DataFrame({
    'Study_Hours':[1,2,3,4,5,6,7,8],
    'Marks':[35,42,50,58,65,73,82,90]
})

sns.regplot(
    x='Study_Hours',
    y='Marks',
    data=df,
    ci=None
)

plt.title("Study Hours vs Marks")
```
