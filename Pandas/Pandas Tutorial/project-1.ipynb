# Project: Employee Salary Analysis

```python
# The below is the sample data o the employees in a company

import pandas as pd
import numpy as np

data = {
    "EmployeeID": [101,102,103,104,105,106,107,108,109,110,111,112],
    "Name": ["John","Alice","Bob","Emma","David","Sophia",
             "Mike","Olivia","James","Linda","Chris","Sarah"],
    "Age": [25,30,28,35,40,26,29,31,45,27,32,38],
    "Department": ["IT","HR","IT","Finance","Manager","HR",
                   "Finance","IT","Manager","HR","Finance","IT"],
    "Experience": [2,5,3,8,15,2,4,6,20,1,7,10],
    "Salary": [45000,55000,50000,70000,120000,48000,
               65000,60000,150000,42000,72000,80000]
}

df = pd.DataFrame(data)
print(df)
```

## Part 1: Data Explorations

```python
# Task 1
print("First 5 rows:")
display(df.head(5))

print("Last 3 rows:")
display(df.tail(3))

print("Shape of the dataset:")
display(df.shape)

print("Columns:")
display(df.columns)
```

```python
# Task 2
print("Data Types:")
display(df.dtypes)

print("Statistics:")
display(df.describe())

print("Null Values:")
display(df.isna())
```

```python
# Task 3
print("Total Employees:")
display(df.shape[0])

print("Salary Stats:")
display(df['Salary'].agg(['mean', 'max', 'min']))
```

## Part 2: Selection & Filtering

```python
# Task 4
print("All employees in IT Department:")
display(df[df['Department'].isin(['IT'])])

print("All employees in HR Department:")
display(df[df['Department'].isin(['HR'])])
```

```python
# Task 5
print("Emplyees Aged above 30:")
age_mask = df['Age'] > 30
display(df[age_mask])

print("Salary above 60000:")
salary_mask = df['Salary'] > 60000
display(df[salary_mask])

print("Experience greater than 5 years:")
experience_mask = df['Experience'] > 5
display(df[experience_mask])
```

```python
# Task 6
print("Employees belonging to IT and Finance:")
display(df[df['Department'].isin(["IT", "Finance"])])
```

```python
# Task 7
print("Salary greater then 70000:")
display(df.query("Salary > 70000"))
```

## Part 4: Feature Engineering

```python
# Task 10
print("create new bonus column:")
df['Bonus'] = df["Salary"] * 0.10
display(df)
```

```python
# Task 11
print("Create new column total pay:")
df['TotalPay'] = df['Salary'] + df['Bonus']
display(df)
```

```python
# Task 12
df["Age_Group"] = df["Age"].map(
    lambda x: "Young" if x < 30 else "Mid" if x <= 40 else "Senior"
)
display(df)
```

## Part 5: Aggregation

```python
# Task 14
print("Department wise employee count:")
display(df.groupby('Department').size())
```

```python
# Task 15
print("Total Salary paid by department:")
display(df.groupby('Department')['Salary'].sum())
```

```python
# Task 16
print('Department with highest salary:')
display(df.groupby('Department')['Salary'].sum().idxmax())
```
