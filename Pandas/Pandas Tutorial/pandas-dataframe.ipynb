# Dataframe

```python
# Creating a dataframe - dataframe is a combinations of multiple series (a series only stores one feature whereas a dataframe holds multiple)
import pandas as pd
import numpy as np

data = {
    "Age": [22, 35, 28, 42],
    "MonthlySpend": [3000, 12000, 5000, 18000],
    "VisitsPerMonth": [15, 30, None, 40],
    "PremiumMember": ["No", "Yes", "No", "Yes"]
}

df = pd.DataFrame(data)

print(df)
```

## In a dataframe colums are the features and the rows are the observtions (In ML if a dataframe has three colums like the area, bedrooms and price then simply the area and bedrooms are features and the price is a target)

## Dataframe data viewing functions

```python
print(df.head(2))
print(df.tail(2))
print(df.info()) # Shows info about rows and columns
print(df.describe()) # Shows statistics (only numeric columns will be shown)
print(df.shape) # Describes the shape of the dataframe (In this case will print (4,4) (rows, columns))
print(df.dtypes) # will show datatypes of each columns
print(df.columns)
```

## In ML a row is a sample while the colums is a feature

```python
# Getting the single column data
print(df['Age'])
print(type(df['Age'])) # This will print pandas.Series as a dataframe column is a series
```

```python
# Getting multiple columns
print(df[['Age', 'MonthlySpend']])
print(type(df[['Age', 'MonthlySpend']])) # This will print pandas.DataFrame because here we are dealing with multiple columns/features together
```

```python
# DataFrame creation functions
# 1 - from_dict() - Creates a dataframe from a dict
data = {
    "Age": [22, 24, 34],
    "Name": ['Ahmer', 'Faraz', 'test user'],
    "Salary": [220000, 300000, 25000]
}

df = pd.DataFrame.from_dict(data)
print(df)

# 2 - from_records() - Creates dataframe from a tuple dataset
records = [
    ("Ahmer", 22),
    ("Faraz", 24) # Here as you can see that each tupple is equal to a row with different column values
]

df = pd.DataFrame.from_records(records, columns=["Name", "Age"]) # In this way we can define column names
print(df)
```

## Column Selection Functions

```python
df = pd.DataFrame({
    "Name": ["Ahmed", "Ali", "Sara", "John", "Zara"],
    "Age": [22, 25, 21, 30, 28],
    "Salary": [50000, 60000, 55000, 80000, 75000],
    "Department": ["IT", "HR", "IT", "Manager", "HR"]
})

print(df)
```

```python
# Get single column
print(df['Age']) # This single bracket [] has a reason here as this will return a single column which will be a series
# Multiple columns
print(df[['Age', 'Name']]) # This double bracket [[]] will return a dataframe
```
