# Dataframe

```python
# Creating a dataframe - dataframe is a combinations of multiple series (a series only stores one feature whereas a dataframe holds multiple)
import pandas as pd
import numpy as np

data = {
    "Age": [22, 35, 28, 42],
    "MonthlySpend": [3000, 12000, 5000, 18000],
    "VisitsPerMonth": [15, 30, None, 40],
    "PremiumMember": ["No", "Yes", "No", "Yes"]
}

df = pd.DataFrame(data)

print(df)
```

## In a dataframe colums are the features and the rows are the observtions (In ML if a dataframe has three colums like the area, bedrooms and price then simply the area and bedrooms are features and the price is a target)

## Dataframe data viewing functions

```python
print(df.head(2))
print(df.tail(2))
print(df.info()) # Shows info about rows and columns
print(df.describe()) # Shows statistics (only numeric columns will be shown)
print(df.shape) # Describes the shape of the dataframe (In this case will print (4,4) (rows, columns))
print(df.dtypes) # will show datatypes of each columns
print(df.columns)
```

## In ML a row is a sample while the colums is a feature

```python
# Getting the single column data
print(df['Age'])
print(type(df['Age'])) # This will print pandas.Series as a dataframe column is a series
```

```python
# Getting multiple columns
print(df[['Age', 'MonthlySpend']])
print(type(df[['Age', 'MonthlySpend']])) # This will print pandas.DataFrame because here we are dealing with multiple columns/features together
```

```python
# DataFrame creation functions
# 1 - from_dict() - Creates a dataframe from a dict
data = {
    "Age": [22, 24, 34],
    "Name": ['Ahmer', 'Faraz', 'test user'],
    "Salary": [220000, 300000, 25000]
}

df = pd.DataFrame.from_dict(data)
print(df)

# 2 - from_records() - Creates dataframe from a tuple dataset
records = [
    ("Ahmer", 22),
    ("Faraz", 24) # Here as you can see that each tupple is equal to a row with different column values
]

df = pd.DataFrame.from_records(records, columns=["Name", "Age"]) # In this way we can define column names
print(df)
```

## Column Selection Functions

```python
df = pd.DataFrame({
    "Name": ["Ahmed", "Ali", "Sara", "John", "Zara"],
    "Age": [22, 25, 21, 30, 28],
    "Salary": [50000, 60000, 55000, 80000, 75000],
    "Department": ["IT", "HR", "IT", "Manager", "HR"]
})

print(df)
```

## Row selection functions

```python
df = pd.DataFrame({
    "Name": ["Ahmed", "Ali", "Sara", "John", "Zara"],
    "Age": [22, 25, 21, 30, 28],
    "Salary": [50000, 60000, 55000, 80000, 75000],
    "Department": ["IT", "HR", "IT", "Manager", "HR"]
})

print(df)
```

```python
# Method 1: loc -label based row selection syntax: df.loc[row_label]
# Selecting single row with loc
print(df.loc[1])

# Selecting multiple rows with loc
print(df.loc[[0, 2]])

# Range selection with loc
print(df.loc[1:3]) # Unlike the python slicing this includes the 3rd value and do not skip it because these are labels

# Selecting specific rows and columns - we can also select specific rows and columns
print(df.loc[1:3, ['Name', 'Age']])
```

```python
# Method 2: iloc - Position based selection Syntax: df.iloc[position]
# Single row selection
print(df.iloc[1]) # This will work regardless of the labels as these are the position or the default indexes this will work even if we define manual labels to the dataframe

# Selecting multiple rows
print(df.iloc[[1,2,3]])

# Range selection
print(df.iloc[1:2]) # As this is the positions this will follow the python slicing rules excluding the nth index or the 2 in this case

# Selecting rows and columns 
print(df.iloc[0, 1]) # Syntax - df.iloc[row, columns]
print(df.iloc[:, 1]) # Get the entire first column
print(df.iloc[:, 0:1]) # This will select all rows but the columns will be only the 0 means first column it will skip the 1 column
```

## Filtering functions

```python
df = pd.DataFrame({
    "Name":["Ahmed","Ali","Sara","John","Zara"],
    "Age":[22,25,21,30,28],
    "Salary":[50000,60000,55000,80000,75000],
    "Department":["IT","HR","IT","Manager","HR"]
})

print(df)
```

```python
# Conditions
print(df['Age'] > 25) # This is called boolean mask (df['Age'] > 25 -> This is the mask)

# Using the mask we can get the actual data the mask only prints true and false
print(df[df['Age'] > 25])
```

```python
# isin() - Used when you want to check whether values belong to a list of allowed values.
df = pd.DataFrame({
    "Name":["Ahmed","Ali","Sara","John","Zara"],
    "Department":["IT","HR","IT","Manager","HR"]
})

print(df[df['Department'].isin(["IT", 'HR'])])

# Opposite of isin() - Selects the row that is not a part of the row
print(df[~df["Department"].isin(["IT","HR"])])
```

```python
# between() - Used when you want a range.

df = pd.DataFrame({
    "Age":[22,25,21,30,28],
    "Experience":[1,3,1,8,5],
    "Salary":[50000,60000,55000,80000,75000]
})

print(df[df['Salary'].between(55000, 70000)])
```

```python
# Unique - Shows all distinct values.
df = pd.DataFrame({
    "Name":["Ahmed","Ali","Sara","John","Zara"],
    "Department":["IT","HR","IT","Manager","HR"]
})

print(df['Department'].unique())
```

```python
# nunique - Counts distinct values.
print(df['Department'].nunique())
```

```python
# value_counts() - Counts occurunces
print(df['Department'].value_counts())

# Normalize - shows percentage
print(df['Department'].value_counts(normalize=True))
```

```python
# Sample() - Randomly selects rows

print(df.sample())

# You can also define how many random rows to select
print(df.sample(2))
```

```python
# query - Alternative filtering syntax.

df = pd.DataFrame({
    "Name":["Ahmed","Ali","Sara","John","Zara"],
    "Age":[22,25,21,30,28],
    "Salary":[50000,60000,55000,80000,75000],
    "Department":["IT","HR","IT","Manager","HR"]
})

print(df.query("Age > 25"))

# Multiple conditions
print(df.query("Age > 25 and Salary > 70000"))
```
