```python
merged_customer_df = pd.merge(merged_customer_df, orders_payment_df, on='order_id', how='left')
merged_customer_df
```

```python
merged_customer_df['payment_installments'].value_counts()
```

```python
def payment_group(installments):
    if installments < 5:
        return 'High instalments Amount'
    elif installments >= 5 and installments <= 12:
        return 'Normal installments Amount'
    else:
        return 'Low installment Amount'

merged_customer_df['InstallmentAmountType'] = merged_customer_df['payment_installments'].apply(payment_group)
merged_customer_df
```
